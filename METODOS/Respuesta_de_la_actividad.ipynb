2.	Realice un histograma para la tasación de los vehículos marca Changan en 10 intervalos de igual amplitud y señale la barra donde se encuentre la moda de color verde. Comente




```python
bdd_changan=datos[datos["MARCA_H"]=="CHANGAN"]#primero se genera la bdd de los vehiculos
bdd_changan
```

```python
bdd_changan.shape #se agrego para saber cuantas observaciones tiene la bdd nueva
```

```python
min = bdd_changan["TASACIÓN"].min()
```

```python
max = bdd_changan["TASACIÓN"].max()+1
```

```python
amplitud = (max-min)/10
```

```python
bind= np.arange(min, max, amplitud)
bind
```

```python
plt.figure(figsize=(10, 6))
n, bins, patches=plt.hist(datos["TASACIÓN"], bins= bind, edgecolor='black', alpha=0.7)
plt.title('Histograma de la Tasación de los vehiculos Changan')
plt.xlabel('Tasación')
plt.ylabel('Cantidad de vehículos')
plt.grid(axis='y', linestyle='--', alpha=0.7)
patches[4].set_facecolor("green") #color asignado al intervalo modal
plt.show()
```

Comentario: Se observa una mayor variabilidad en los precios más bajos de las tasaciones de los vehiculos Changan.
Existe menos variabilidad en los precios más altos de las tasaciones de los vehiculos Changan.
